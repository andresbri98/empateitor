Vamos a intentar rellenar nulos de las cuotas, ya que son muy importantes. La casa de apuestas es indiferente, ya que todas tendrás cuotas similares. La idea es quedarnos solo con una casa de apuestas y rellenar los nulos de esas columnas con otras cuotas quesí vengan informados, priorizaremos la casa de apuestas B365

```python
# Vista previa de cambios en cuotas (antes de aplicar prioridad B365)
import numpy as np
import pandas as pd

houses = {
    'H': {'primary': 'B365H', 'others': ['BWH', 'WHH', 'VCH', 'IWH']},
    'D': {'primary': 'B365D', 'others': ['BWD', 'WHD', 'VCD', 'IWD']},
    'A': {'primary': 'B365A', 'others': ['BWA', 'WHA', 'VCA', 'IWA']},
}

def to_numeric_if_exists(df, col):
    if col in df.columns:
        return pd.to_numeric(df[col], errors='coerce')
    return pd.Series(np.nan, index=df.index)

def coalesce_with_primary(df, primary, others):
    base = to_numeric_if_exists(df, primary)
    out = base.copy() if base is not None else pd.Series(np.nan, index=df.index)
    for c in others:
        s = to_numeric_if_exists(df, c)
        out = out.fillna(s)
    return out

def first_nonnull_source_with_primary(df, primary, others):
    candidates = [primary] + others
    src = pd.Series(None, index=df.index, dtype=object)
    for c in candidates:
        s = to_numeric_if_exists(df, c)
        mask = s.notna()
        src = src.mask(mask & src.isna(), c)
    return src

def build_preview_for(outcome):
    primary = houses[outcome]['primary']
    others = houses[outcome]['others']
    before = to_numeric_if_exists(df_simplificado, primary)
    after = coalesce_with_primary(df_simplificado, primary, others)
    src = first_nonnull_source_with_primary(df_simplificado, primary, others)
    changed = ((before.isna() & after.notna()) | ((before.notna()) & (after.notna()) & (before != after)))
    cols_ctx = [c for c in ['Season','Competition','Date','HomeTeam','AwayTeam'] if c in df_simplificado.columns]
    preview = df_simplificado.loc[changed, cols_ctx].copy()
    preview['Outcome'] = outcome
    preview[f'BEFORE_{outcome}'] = before[changed]
    preview[f'AFTER_{outcome}'] = after[changed]
    preview[f'SOURCE_{outcome}'] = src[changed]
    return preview.head(10)

preview_frames = []
for oc in ['H','D','A']:
    preview_frames.append(build_preview_for(oc))
preview_changes = pd.concat(preview_frames, ignore_index=True)
display(preview_changes)
```

```python
# Relleno de nulos en cuotas priorizando Bet365
import numpy as np
import pandas as pd

# Conjunto de columnas por resultado
houses = {
    'H': {
        'primary': 'B365H',
        'others': ['BWH', 'WHH', 'VCH', 'IWH']
    },
    'D': {
        'primary': 'B365D',
        'others': ['BWD', 'WHD', 'VCD', 'IWD']
    },
    'A': {
        'primary': 'B365A',
        'others': ['BWA', 'WHA', 'VCA', 'IWA']
    }
}

# Función: convierte columnas a numérico (si existen)
def to_numeric_if_exists(df, col):
    if col in df.columns:
        return pd.to_numeric(df[col], errors='coerce')
    return pd.Series(np.nan, index=df.index)

# Función: coalesce priorizando 'primary'; si está nulo, toma la primera disponible de 'others'.
# Nota: el orden de 'others' es arbitrario (sin prioridad específica).
def coalesce_with_primary(df, primary, others):
    base = to_numeric_if_exists(df, primary)
    out = base.copy() if base is not None else pd.Series(np.nan, index=df.index)
    for c in others:
        s = to_numeric_if_exists(df, c)
        out = out.fillna(s)
    return out

# Fuente de la primera cuota no nula usada

def first_nonnull_source_with_primary(df, primary, others):
    candidates = [primary] + others
    src = pd.Series(None, index=df.index, dtype=object)
    for c in candidates:
        s = to_numeric_if_exists(df, c)
        mask = s.notna()
        src = src.mask(mask & src.isna(), c)
    return src

# Construir ODDS_* y sus fuentes priorizando Bet365
for outcome, cfg in houses.items():
    primary = cfg['primary']
    others = cfg['others']
    df_simplificado[f'ODDS_{outcome}'] = coalesce_with_primary(df_simplificado, primary, others)
    df_simplificado[f'ODDS_SOURCE_{outcome}'] = first_nonnull_source_with_primary(df_simplificado, primary, others)

# Rellenar específicamente las columnas Bet365 cuando estén nulas
for outcome, cfg in houses.items():
    primary = cfg['primary']
    if primary in df_simplificado.columns:
        df_simplificado[primary] = to_numeric_if_exists(df_simplificado, primary)
        df_simplificado[primary] = df_simplificado[primary].fillna(df_simplificado[f'ODDS_{outcome}'])

# Resumen de nulos y cobertura
cols_report = ['B365H', 'B365D', 'B365A', 'BWH', 'BWD', 'BWA', 'WHH', 'WHD', 'WHA', 'VCH', 'VCD', 'VCA', 'IWH', 'IWD', 'IWA',
               'ODDS_H', 'ODDS_D', 'ODDS_A', 'ODDS_SOURCE_H', 'ODDS_SOURCE_D', 'ODDS_SOURCE_A']
existentes = [c for c in cols_report if c in df_simplificado.columns]

nulls = df_simplificado[existentes].isna().sum().sort_index()
coverage = (1 - df_simplificado[['ODDS_H','ODDS_D','ODDS_A']].isna().mean()) * 100

print('Nulos por columna (tras relleno con prioridad B365):')
print(nulls)
print('\nCobertura ODDS_% (no nulos) con prioridad B365:')
print(coverage.round(2))

# Top fuente usada para ODDS_H
if 'ODDS_SOURCE_H' in df_simplificado.columns:
    top_src = df_simplificado['ODDS_SOURCE_H'].value_counts(dropna=True).head(5)
    print('\nTop fuentes para ODDS_H:')
    print(top_src)

# Eliminar columnas de casas de apuestas y fuentes, quedarnos solo con ODDS
cols_houses = ['B365H','B365D','B365A','BWH','BWD','BWA','WHH','WHD','WHA','VCH','VCD','VCA','IWH','IWD','IWA']
cols_sources = ['ODDS_SOURCE_H','ODDS_SOURCE_D','ODDS_SOURCE_A']
cols_to_drop = [c for c in (cols_houses + cols_sources) if c in df_simplificado.columns]

if cols_to_drop:
    df_simplificado.drop(columns=cols_to_drop, inplace=True)

print('\nColumnas finales en df_simplificado:')
print(list(df_simplificado.columns))
```

```python
# Muestra aleatoria de La Liga (temporadas recientes)
import numpy as np
import pandas as pd

# Helper: convertir Season a año base (primer año de la temporada)
def season_to_year(season):
    s = str(season)
    # Buscar año de 4 dígitos
    import re
    m4 = re.search(r"(19|20)\d{2}", s)
    if m4:
        return int(m4.group(0))
    # Buscar dos dígitos
    m2 = re.search(r"\b(\d{2})\b", s)
    if m2:
        yy = int(m2.group(1))
        return 1900 + yy if yy >= 90 else 2000 + yy
    return np.nan

# Filtros por competición y división
comp_ok = df_simplificado['Competition'].astype(str).str.lower().str.contains('la liga|laliga|la_liga', regex=True) if 'Competition' in df_simplificado.columns else pd.Series(False, index=df_simplificado.index)
div_ok = df_simplificado['Div'].astype(str).str.upper().isin(['SP1']) if 'Div' in df_simplificado.columns else pd.Series(False, index=df_simplificado.index)

years = df_simplificado['Season'].apply(season_to_year) if 'Season' in df_simplificado.columns else pd.Series(np.nan, index=df_simplificado.index)
recent_ok = years >= 2018

mask = (comp_ok | div_ok) & recent_ok
subset = df_simplificado.loc[mask].copy()

if subset.empty:
    print('No se encontraron partidos de La Liga en temporadas recientes (>=2018).')
else:
    cols_view = [c for c in ['Season','Date','Competition','Div','HomeTeam','AwayTeam','FTHG','FTAG','FTR','ODDS_H','ODDS_D','ODDS_A'] if c in subset.columns]
    n = min(10, len(subset))
    rng_state = np.random.RandomState()
    sample = subset.sample(n, random_state=rng_state.randint(0, 1_000_000))
    display(sample[cols_view])
    # Mini-resumen de cobertura de cuotas en el subset
    cov = (1 - sample[['ODDS_H','ODDS_D','ODDS_A']].isna().mean()) * 100
    print('\nCobertura de cuotas en la muestra (% no nulos):')
    print(cov.round(2))
```

Ahora vamos a analizar los partidos de los que no tenemos resultados. Es decir, variables no informadas en FTAG, FTHG y FTR

```python
# Analizar y eliminar grupos con resultados no informados (FTAG, FTHG, FTR)
import numpy as np
import pandas as pd

# Preparar chequeos robustos para FTHG, FTAG y FTR
FTHG_num = pd.to_numeric(df_simplificado['FTHG'], errors='coerce') if 'FTHG' in df_simplificado.columns else pd.Series(np.nan, index=df_simplificado.index)
FTAG_num = pd.to_numeric(df_simplificado['FTAG'], errors='coerce') if 'FTAG' in df_simplificado.columns else pd.Series(np.nan, index=df_simplificado.index)

if 'FTR' in df_simplificado.columns:
    FTR_clean = df_simplificado['FTR'].astype(str).str.strip()
    FTR_clean = FTR_clean.replace({'': np.nan})
    FTR_valid = FTR_clean.str.upper().where(FTR_clean.str.upper().isin(['H','D','A']))
else:
    FTR_valid = pd.Series(np.nan, index=df_simplificado.index)

# Un partido está "no informado" si falta cualquiera de FTHG, FTAG o FTR
uninformed = (
    FTHG_num.isna() |
    FTAG_num.isna() |
    FTR_valid.isna()
)

# Elegir columnas de agrupación (Division preferida; si no, Div)
group_cols = []
if 'Division' in df_simplificado.columns:
    group_cols.append('Division')
elif 'Div' in df_simplificado.columns:
    group_cols.append('Div')
if 'Competition' in df_simplificado.columns:
    group_cols.append('Competition')
if 'Season' in df_simplificado.columns:
    group_cols.append('Season')

if not group_cols:
    print('No hay columnas adecuadas para agrupar (Division/Div, Competition, Season).')
else:
    # Añadir bandera al dataframe para agregaciones
    df_simplificado['__uninformed'] = uninformed

    # Resumen por grupo: total y no informados
    summary = df_simplificado.groupby(group_cols, dropna=False)['__uninformed'].agg(total='size', uninformed_count='sum')
    summary['uninformed_ratio'] = (summary['uninformed_count'] / summary['total']).round(4)

    # Mostrar únicamente los grupos con algún partido no informado
    summary_to_display = summary[summary['uninformed_count'] > 0].sort_values(['uninformed_ratio','uninformed_count'], ascending=False)
    print('Resumen de grupos con partidos no informados (ordenados por ratio y cantidad):')
    display(summary_to_display)

    # También mostrar ejemplos de registros no informados por grupo (vista compacta)
    cols_ctx = [c for c in ['Season','Competition','Division','Div','Date','HomeTeam','AwayTeam','FTHG','FTAG','FTR'] if c in df_simplificado.columns]
    sample_uninformed = df_simplificado.loc[df_simplificado['__uninformed'], cols_ctx].copy()
    print('\nMuestra (hasta 30 filas) de partidos no informados:')
    display(sample_uninformed.head(30))

    # Umbral: eliminar grupos que superen el 10% de partidos no informados
    ratio_threshold = 0.10
    high_missing_groups = summary[summary['uninformed_ratio'] > ratio_threshold].reset_index()

    # Construir máscara de filas a eliminar usando clave compuesta robusta (manejo de NaN)
    def build_composite_key(df_like, cols):
        if df_like.empty:
            return pd.Series([], dtype=object)
        tmp = df_like[cols].copy()
        for c in cols:
            tmp[c] = tmp[c].astype(str).where(tmp[c].notna(), '<<NA>>')
        return tmp.apply(lambda r: tuple(r.values.tolist()), axis=1)

    keys_df = build_composite_key(df_simplificado, group_cols)
    keys_remove = set(build_composite_key(high_missing_groups, group_cols))
    remove_mask = keys_df.isin(keys_remove)

    # Reporte de grupos eliminados
    removed_groups_report = []
    if not high_missing_groups.empty:
        for _, row in high_missing_groups.iterrows():
            removed_groups_report.append({
                'reason': '>10% no informados',
                'Season': row.get('Season', None),
                'Competition': row.get('Competition', None),
                'Division': row.get('Division', row.get('Div', None)),
                'total': int(row['total']),
                'uninformed_count': int(row['uninformed_count']),
                'uninformed_ratio': float(row['uninformed_ratio'])
            })

    # Aplicar borrados
    before_rows = len(df_simplificado)
    if remove_mask.any():
        df_simplificado = df_simplificado.loc[~remove_mask].copy()
        print(f"\nSe eliminaron {remove_mask.sum()} filas de grupos con >10% de partidos no informados.")
    else:
        print('\nNo se eliminaron filas: no hay grupos con >10% no informados.')

    # Mostrar qué Season y Competition se han borrado
    if removed_groups_report:
        removed_df = pd.DataFrame(removed_groups_report)
        print('\nGrupos eliminados (Season, Competition, Division y motivo):')
        display(removed_df[['reason','Season','Competition','Division','total','uninformed_count','uninformed_ratio']])
        # Listado simple solicitado (Season y Competition)
        removed_pairs = removed_df[['Season','Competition']].drop_duplicates()
        print('\nSeason y Competition eliminados por no estar informados (>10%):')
        display(removed_pairs)
    else:
        print('\nNo se eliminaron grupos completos por falta de información (>10%).')

    # Limpiar columnas auxiliares si existiesen en el dataframe resultante
    aux_cols = [c for c in ['__uninformed'] if c in df_simplificado.columns]
    if aux_cols:
        df_simplificado.drop(columns=aux_cols, inplace=True)

    # Resumen final
    after_rows = len(df_simplificado)
    print(f"\nFilas antes: {before_rows} | Filas después: {after_rows}")
    print('Columnas actuales en df_simplificado:')
    print(list(df_simplificado.columns))
```

Dado que son todos de 4 temporadas de la misma liga, eliminamos por completo toda esa competicion durante esas temporadas

```python
# Eliminar registros de superliga_griega en seasons específicas
import pandas as pd

# Partros de borrado
TARGET_COMPETITION = 'superliga_griega'
TARGET_SEASONS = ['00-01', '02-03', '03-04', '04-05']

# Validaciones de columnas
missing_cols = [c for c in ['Competition','Season'] if c not in df_simplificado.columns]
if missing_cols:
    print(f"Faltan columnas requeridas en df_simplificado: {missing_cols}. No se puede proceder.")
else:
    comp_vals = df_simplificado['Competition'].astype(str).str.strip().str.lower()
    season_vals = df_simplificado['Season'].astype(str).str.strip()

    comp_mask = comp_vals.eq(TARGET_COMPETITION)
    season_mask = season_vals.isin(TARGET_SEASONS)
    remove_mask = comp_mask & season_mask

    before_rows = len(df_simplificado)
    to_remove = int(remove_mask.sum())

    # Resumen de eliminados por Season (y Division si existe)
    cols_group = ['Season'] + ([ 'Division' ] if 'Division' in df_simplificado.columns else ([ 'Div' ] if 'Div' in df_simplificado.columns else []))
    removed_summary = (
        df_simplificado.loc[remove_mask, cols_group]
        .groupby(cols_group, dropna=False)
        .size()
        .to_frame('removed_rows')
        .sort_values('removed_rows', ascending=False)
    )

    # Aplicar borrado
    df_simplificado = df_simplificado.loc[~remove_mask].copy()
    after_rows = len(df_simplificado)

    print(f"Filas antes: {before_rows}")
    print(f"Filas eliminadas (superliga_griega {TARGET_SEASONS}): {to_remove}")
    print(f"Filas después: {after_rows}")

    if to_remove > 0:
        print('\nResumen de filas eliminadas por grupo (Season y Division/Div):')
        display(removed_summary)

    # Lista única de seasons eliminadas (por trazabilidad)
    removed_seasons = sorted(set(season_vals[remove_mask].tolist()))
    print('\nSeasons afectadas eliminadas:')
    print(removed_seasons)
```

A las variables AR, AY, HR y HY dado que son tarjetas, vamos a asignar el valor de la moda.
Para las nulos en HTAG, HTHG y HR, aunque no sea del todo correcto y estemos suponiendo que el partido acabó igual que iba al descanso, vamos a introducir el valor de FTAG, FTHG y FTR

```python
# Rellenar tarjetas con moda y completar HTAG/HTHG/HTR desde FTAG/FTHG/FTR
import numpy as np
import pandas as pd

cols_cards = ['AR','AY','HR','HY']
cols_half_final_map = {
    'HTAG': 'FTAG',
    'HTHG': 'FTHG',
    'HTR':  'FTR',
}

# Resumen antes
before_nulls = {}
for c in cols_cards + list(cols_half_final_map.keys()):
    if c in df_simplificado.columns:
        before_nulls[c] = int(df_simplificado[c].isna().sum())

# 1) Rellenar tarjetas con la moda global de cada columna
for c in cols_cards:
    if c in df_simplificado.columns:
        s = df_simplificado[c]
        # Convertir a numérico cuando tiene forma numérica
        s_num = pd.to_numeric(s, errors='ignore')
        mode_vals = pd.Series(s_num).mode(dropna=True)
        if len(mode_vals) > 0:
            mode_val = mode_vals.iloc[0]
            df_simplificado[c] = s_num.fillna(mode_val)

# 2) Rellenar HTAG/HTHG/HTR a partir de FTAG/FTHG/FTR
for half_col, final_col in cols_half_final_map.items():
    if half_col in df_simplificado.columns and final_col in df_simplificado.columns:
        if half_col in ['HTAG','HTHG']:
            half = pd.to_numeric(df_simplificado[half_col], errors='coerce')
            final = pd.to_numeric(df_simplificado[final_col], errors='coerce')
            df_simplificado[half_col] = half.fillna(final)
        else:  # HTR desde FTR
            half = df_simplificado[half_col].astype(str).str.strip()
            half = half.replace({'': np.nan})
            final = df_simplificado[final_col].astype(str).str.strip().str.upper()
            final = final.replace({'': np.nan})
            # Sólo considerar FTR válido H/D/A
            final_valid = final.where(final.isin(['H','D','A']))
            df_simplificado[half_col] = half.where(half.notna(), final_valid)

# Resumen después
after_nulls = {}
for c in cols_cards + list(cols_half_final_map.keys()):
    if c in df_simplificado.columns:
        after_nulls[c] = int(df_simplificado[c].isna().sum())

print('Nulos antes -> después (columnas clave):')
for c in sorted(set(list(before_nulls.keys()) + list(after_nulls.keys()))):
    print(f"{c}: {before_nulls.get(c, 'NA')} -> {after_nulls.get(c, 'NA')}")

print('\nColumnas actuales en df_simplificado:')
print(list(df_simplificado.columns))
```
